There are many ways to handle null values.

For example, for missing numerical values you can use the mean of that variable to fill in the missing values.

```python
fifa_data_df.fillna(fifa_data_df.mean(), 
                    inplace=True)
```

Now that we have cleaned the data, we can begin the exploratory data analysis phase.

```python
fifa_data_df[['Name', 'Club', 'Position', 'Value']]
```

From the above output I can see that Value is a object (string) based variable this is due to the M or K indicating millions or thousands.

Let's create a new variable which has a numeric data type.
From the above output we can see that we need to remove the currency symbol and remove the end character if it is M or K

```python
fifa_data_df['numeric_value'] = fifa_data_df['Value'].str.split('€').str[-1]
```

```python
fifa_data_df['numeric_value'] = np.where(fifa_data_df['numeric_value'].str.strip().str[-1] == 'M', 
                                         fifa_data_df['numeric_value'].str[0:-1], 
                                         fifa_data_df['numeric_value'])
fifa_data_df['numeric_value'] = np.where(fifa_data_df['numeric_value'].str.strip().str[-1] == 'K', 
                                         fifa_data_df['numeric_value'].str[0:-1], 
                                         fifa_data_df['numeric_value'])
fifa_data_df['numeric_value'] = fifa_data_df['numeric_value'].astype(float)
```

```python
fifa_data_df['numeric_value'] = np.where(fifa_data_df['Value'].str[-1] == 'M', 
                                         fifa_data_df['numeric_value'] * (1 * (10**6)),
                                         fifa_data_df['numeric_value'])
fifa_data_df['numeric_value'] = np.where(fifa_data_df['Value'].str[-1] == 'K', 
                                         fifa_data_df['numeric_value'] * (1 * (10**3)),
                                         fifa_data_df['numeric_value'])
```

## Data visualisations

```python
# comparison of preferred foot over the different players

plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(fifa_data_df['Preferred Foot'], 
              palette = 'Pastel1')
plt.title('Most Preferred Foot of the Players', 
          fontsize = 20)
plt.show()
```

```python
# To show Different nations participating in the FIFA 2018

plt.style.use('dark_background')
fifa_data_df['Nationality'].value_counts().head(80).plot.bar(color = 'orange', figsize = (20, 7))
plt.title('Different Nations Participating in FIFA 2018', fontsize = 30, fontweight = 20)
plt.xlabel('Name of The Country')
plt.ylabel('count')
plt.show()
```

Now let's look at the distribution of age

```python
plt.style.use('dark_background')
fifa_data_df.groupby('Age')['Age'].size().plot.bar(color = 'orange', figsize = (20, 7))
plt.title('Distribution of players age in 2018', fontsize = 30, fontweight = 20)
plt.ylabel('count')
plt.show()
```

It could be interesting to understand the relationship between age and numeric value

```python
positions_df = fifa_data_df[['Position', 'Age', 'numeric_value']]
positions_df = positions_df.groupby('Position')

plt.rcParams['figure.figsize'] = (20, 10)
for name, group in positions_df:
    plt.plot(group.Age, 
             group.numeric_value, 
             marker='.', 
             markersize=12,
             linestyle='',
             label=name)
plt.legend()  
plt.ticklabel_format(style='plain')
plt.ylabel('Value in euros')
plt.xlabel('Age')
plt.show()
```

Let's take a look at the mean player features for each team and position

```python
feature_cols = ['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 
                'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 
                'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 
                'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 
                'GKKicking', 'GKPositioning', 'GKReflexes']

club_pos_mean_features_df = fifa_data_df.groupby(['Club', 'Position'])[feature_cols].agg('mean')
club_pos_mean_features_df.tail()
```

```python
club_pos_mean_features_df.sort_values(by='Crossing', 
                                      ascending=False).head(20)
```

```python
club_pos_mean_features_df[['Agility', 'Dribbling']]
```

```python
ag_vs_dr_df = club_pos_mean_features_df[['Agility', 'Dribbling']]
ag_vs_dr_df = ag_vs_dr_df.sort_values('Position', ascending=False).head(25)
ag_vs_dr_df = ag_vs_dr_df.groupby(['Club', 'Position'])

plt.rcParams['figure.figsize'] = (20, 10)
for name, group in ag_vs_dr_df:
    plt.plot(group.Agility, 
             group.Dribbling, 
             marker='.', 
             markersize=12,
             linestyle='',
             label=name)
plt.legend()  
plt.ticklabel_format(style='plain')
plt.ylabel('Dribbling')
plt.xlabel('Agility')
plt.show()
```
